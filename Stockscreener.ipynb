```python
import pandas as pd
```

```python
df = pd.read_csv("stocks.csv")
df.head()
```

```python
# clean up last sales
df['Last Sale'] = df['Last Sale'].str.replace('$', '').astype(float)
```

```python
# keep stocks with trading volume over 5,000,000
df = df[df['Volume'] > 3000000]
# keep companies with market cap over 2 billion
df = df[df['Market Cap'] > 2000000000]
# remove penny stocks
df = df[df['Last Sale'] > 5]
len(df)
```

```python
# check to see the amount of companies for each sector
sectors = df['Sector'].unique()
for i in sectors:
  sector = df[df['Sector'] == i]
  print(f'The number of {i} tickers is {len(sector)}')
```

```python
# sort the csv by sector
df = df.sort_values(by=['Sector'])
```

```python
df.to_csv('screened_stocks.csv', index=False)
```
